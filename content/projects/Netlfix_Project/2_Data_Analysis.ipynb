## Notebook 2: Data Analysis (`2_Data_Analysis.ipynb`)

**Purpose:** To perform Exploratory Data Analysis (EDA) and generate summary statistics to understand the dataset's composition.

```python
import pandas as pd
df = pd.read_csv('/Users/apple/Documents/Uni/Personal /Netlfix_Project/Data/netflix_cleaned.csv')
df
```

*   **Content Split:** We calculate the ratio of Movies vs. TV Shows to see which content type dominates the platform.


```python
# Ratio of Movies vs TV Shows
type_counts = df['type'].value_counts()
print("Content Types:\n", type_counts)

# Calculate percentage
type_ratios = df['type'].value_counts(normalize=True) * 100
print("\nContent Ratios (%):\n", type_ratios)
```

*   **Geographic Analysis:** By counting unique values in the `country` column, we identify the top content-producing nations on Netflix.

```python
# Top 10 countries producing content
top_countries = df['country'].value_counts().head(10)
print("\nTop 10 Countries:\n", top_countries)
```

*   **Rating Analysis:** We analyze the `rating` column (e.g., PG-13, TV-MA) to understand the target audience demographics (e.g., how much content is for adults vs. children).

```python
# Most common ratings
rating_counts = df['rating'].value_counts()
print("\nRatings Distribution:\n", rating_counts)
```
