```python
import pandas as pd 
import numpy as np
import os
```

```python
RAW_DATA_PATH = "Data/Raw/spotify_top_music.csv"
PROCESSED_DATA_PATH = "Data/spotify_top_music_cleaned.csv"
```

```python
os.makedirs(os.path.dirname(PROCESSED_DATA_PATH), exist_ok=True)
```

```python
df = pd.read_csv(RAW_DATA_PATH, encoding='ISO-8859-1')

print(f"orignal shape: {df.shape}")
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
df = df.drop_duplicates(subset=['title', 'artist'])

```

```python
df_clean = df[df['bpm'] > 0].copy()
```

```python
if df_clean.isnull().sum().sum() > 0:
    print("Found missing values. Filling with median or dropping...")
    df_clean = df_clean.dropna()
print(f"Cleaned shape: {df_clean.shape}")
```

```python
df_clean['duration_min'] = df_clean['dur'] / 60
```

```python
df_clean.to_csv(PROCESSED_DATA_PATH, index=False)
print(f"Cleaned data saved to {PROCESSED_DATA_PATH}")
```
